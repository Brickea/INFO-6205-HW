```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
cost = pd.read_csv("./hw3.csv")
cost
```

```python
x = np.logspace(0,4,5,base=2)*100
x
```

```python
plt.scatter(cost["n (number of sites)"],cost["m (number of connections)"],label="c (number of connections)")
plt.scatter(cost["n (number of sites)"],cost["rp (number of randomly generated pairs)"],label="m (number of randomly generated pairs)")
plt.xlabel("The sites number")

plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
x = (cost["n (number of sites)"]).to_numpy().reshape(-1,1)
```

```python
X = x
```

```python
y = cost["rp (number of randomly generated pairs)"]
```

```python
lin_reg.fit(X, y)
y_predict = lin_reg.predict(X)
plt.scatter(x, cost["rp (number of randomly generated pairs)"])
plt.plot(x, y_predict, color='r')
plt.show()
```
